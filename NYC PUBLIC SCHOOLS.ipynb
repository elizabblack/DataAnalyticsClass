```python
df2=df[['School DBN','Regents Exam','Year','Total Tested', 'Mean Score']]
```

```python
# To seperate out the District 84 schools. 
df5=df4.loc[df4['Borough'] == '84']
```

```python
df5['Total Tested']=df5['Total Tested'].astype(int)
```

```python
## To get the list of schools in District 84 that had a Total Tested of 300+.
df6=df5.loc[df5['Total Tested'] >= 300]
```
